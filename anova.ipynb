```python
import pandas as pd
df = pd.read_csv('ANOVA.csv')
df
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols('value ~ C(treatment)', data=df).fit()
anova_table = sm.stats.anova_lm(model,typ=2)
anova_table
```

```python
from bioinfokit.analys import stat
res = stat()
res.anova_stat(data=df, res_var='value', anova_model='value ~ C(treatments)')
res.anova_summary
```

```python
from bioinfokit.analys import stat
res =stat()
res.tukey_hsd(df, res_var='value',xfac_var='tratments', anova_model='value ~ C(treatment)')
res.tukey_summary
```
