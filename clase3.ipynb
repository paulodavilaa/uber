```python
import numpy as np
import pandas as pd
from datetime import date, time, datetime, timedelta
import math

S1 = pd.Series(['Juan', 'Karla', 'Judith'], index=['id1','id2','id3'])
S2 = pd.Series([23,28,25], index=['id1','id2','id3'])
S3 = pd.Series([70.5,56.7,51.2], index=['id1','id2','id3'])
S4 = pd.Series([date(1997,10,4),date(2000,3,17),date(2002,7,23)], index=['id1','id2','id3'])
d1 = {'nombre': S1, 'edad': S2, 'peso': S3, 'fnac':S4}
df1 = pd.DataFrame(d1)
for i in range(3):
  nombre = df1.iloc[i,0]
  edadDT = date.today() - df1.iloc[i,3]
  edad = math.floor(edadDT.days/365)
  print(f"La edad de {nombre} es de {edad} años")
df1
lista = ['id1','id2','id3']
for i in lista:
  nombre = df1.loc[i,'nombre']
  edadDT = date.today() - df1.loc[i,'fnac']
  edad = math.floor(edadDT.days/365)
  print(f"La edad de {nombre} es de {edad} años")



df5 = df1.copy()
df5 = pd.DataFrame(df5,index=['id1', 'id2', 'id3', 'id4'])
df5.loc['id4','nombre'] = 'Pedro'
df5
```
