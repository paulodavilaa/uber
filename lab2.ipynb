```python
import pandas as pd
import numpy as np
import io
import requests
import matplotlib.pyplot as plt
%matplotlib inline 
from scipy.stats import norm
import seaborn as sns
from scipy.stats import stats
from scipy.stats import linregress

df_url = 'https://raw.githubusercontent.com/akmand/datasets/master/openintro/bdims.csv'
url_content = requests.get(df_url, verify=False).content
df = pd.read_csv(io.StringIO(url_content.decode('utf-8')))
df.shape
```

```python
df
```

```python
fig, ax=plt.subplots(figsize =(12,10))

#DEFINIMOS x y y
ax.scatter(df['hgt'],df['wgt'])
ax.set_xlabel(('Height'))
ax.set_ylabel(('Weight'))
ax.set_title(('Dispersión peso vs altura'))
```

```python
x=df['hgt'] 
y=df['wgt']

slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

print("R-squared: %f" % r_value**2)
print("P-value: %f" % p_value)

plt.plot(x, y, 'o', label='original data')

plt.plot(x, intercept + slope*x, 'r', label='fitted line')
plt.legend()
plt.show()
print("y = " + str(round(slope,2)) + "x + " + str(round(intercept,2)))
```

```python
coeficiente, pvalue = stats.pearsonr(x, y)
print("Coeficiente de correlación: ", coeficiente)
print("P-value: ", pvalue)
```
