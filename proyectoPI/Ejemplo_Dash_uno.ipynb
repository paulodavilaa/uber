# Se dan de alta las librerías a utilizar

```python
import dash
import dash_html_components as html
import dash_core_components as dcc
import plotly.graph_objects as go
import plotly.express as px
from dash.dependencies import Input, Output
from jupyter_dash import JupyterDash
import pandas as pd
```

# Se inicializa una variable para crear la aplicación

```python
app = JupyterDash(__name__)
```

# Se carga la base de datos

```python
df = pd.read_csv('internacional.csv')
df
```

# Se genera el contenido que se tendrá en la aplicación web

```python
app.layout = html.Div(id = 'parent', children = [
    html.H1(id = 'H1', children = 'Styling using html components', style = {'textAlign':'center',\
                                            'marginTop':40,'marginBottom':40}),

        dcc.Dropdown( id = 'dropdown',
        options = [
            {'label':'Promedio', 'value':'Promedio' },
            ],
        value = 'Promedio'),
        dcc.Graph(id = 'bar_plot')
    ])
```

```python
app.layout = html.Div(id = 'parent', children = [
    html.H1(id = 'H1', children = 'Styling using html components', style = {'textAlign':'center',\
                                            'marginTop':40,'marginBottom':40}),

        dcc.Dropdown( id = 'dropdown',
        options = [
            {'label':'Promedio', 'value':'Promedio' },
            {'label':'Nivel', 'value':'Nivel' },
            ],
        value = 'Promedio'),
        dcc.Graph(id = 'bar_plot')
    ])
```

# Se establece el proceso para la entrada y salida de datos

Se define una función para crear y actualizar el gráfico

```python
@app.callback(Output(component_id='bar_plot', component_property= 'figure'),
              [Input(component_id='dropdown', component_property= 'value')])
def graph_update(dropdown_value):
    print(dropdown_value)
    fig = go.Figure([go.Scatter(x = df['Fecha'], y = df['{}'.format(dropdown_value)],\
                     line = dict(color = 'firebrick', width = 4))
                     ])
    
    fig.update_layout(title = 'Stock prices over time',
                      xaxis_title = 'Fecha',
                      yaxis_title = 'Promedio'
                      )
    return fig
```

```python
@app.callback(Output(component_id='bar_plot', component_property= 'figure'),
              [Input(component_id='dropdown', component_property= 'value')])
def graph_update(dropdown_value):
    print(dropdown_value)
    fig = go.Figure([go.Scatter(x = df['Estatus'], y = df['{}'.format(dropdown_value)],\
                     line = dict(color = 'firebrick', width = 4))
                     ])
    
    fig.update_layout(title = 'Stock prices over time',
                      xaxis_title = 'Estatus',
                      yaxis_title = 'Nivel'
                      )
    return fig
```

# Se ejecuta la aplicación

```python
app.run_server(mode='inline')
```
