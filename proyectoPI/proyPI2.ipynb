```python
def tomarPrimeraOpcion(text):
    text = text[text.find("1 - ")+4:]
    if text.find(",") == -1:
        return text
    else:
        return text[:text.find(",")]
```

```python
df["PrimeraOpción"] = df["OportunidadesSeleccionadas"].apply(tomarPrimeraOpcion)
```

```python
df = df.astype(str)
```
